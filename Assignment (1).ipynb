1. Key differences between SQL and NoSQL databases

- SQL uses structured tables with a fixed schema and uses standardized SQL for queries. NoSQL (like MongoDB) uses flexible document/key-value/graph models, has a dynamic schema, and scales horizontally. SQL follows ACID; NoSQL typically follows BASE consistency.

2. What makes MongoDB a good choice for modern applications?

- MongoDB stores data as JSON-like documents, supports flexible schemas, scales horizontally via sharding, handles nested/varied data naturally, and offers a rich aggregation pipeline — making it ideal for rapidly evolving, high-volume applications.

3. Concept of collections in MongoDB

- A collection is MongoDB's equivalent of a SQL table — a grouping of documents within a database. Unlike tables, collections don't enforce a fixed schema, so documents within the same collection can have different fields and data types.

4. How MongoDB ensures high availability using replication

- MongoDB uses a Replica Set — a group of mongod instances that maintain the same data. One node is the primary (handles all writes), and the rest are secondaries (replicate data from primary). If the primary fails, an automatic election promotes a secondary to primary, ensuring near-zero downtime.

5. Main benefits of MongoDB Atlas

- Atlas is MongoDB's fully managed cloud service. Key benefits: automated backups and point-in-time recovery, built-in monitoring and performance alerts, auto-scaling, multi-cloud/multi-region deployment, built-in security (encryption, VPC peering), and no infrastructure management overhead.

6. Role of indexes in MongoDB and how they improve performance

- Indexes are data structures (B-tree by default) that store a small portion of a collection's data in an easy-to-traverse form. Without an index, MongoDB does a full collection scan (COLLSCAN). With an index, it jumps directly to matching documents (IXSCAN), drastically reducing query time — especially on large collections. Common types: single field, compound, text, geospatial, and TTL indexes.

7. Stages of the MongoDB aggregation pipeline

- $match: filter documents
- $group: Group and Aggregate
- $project: Reshape/ select fields
- $sort: sort results
- $limit/$skip: Paginate result
- $lookup: Join with another collection
- $unwind:  Deconstruct arrays
- $addfields:  Add computed fields


8. Sharding vs Replication in MongoDB
- Replication is about availability — copies of the same data on multiple nodes so the system survives failures. Sharding is about scalability — splitting data across multiple nodes (shards) so no single machine holds all the data. Replication duplicates; sharding distributes.

9. What is PyMongo and why is it used?
- PyMongo is the official Python driver for MongoDB. It allows Python applications to connect to a MongoDB instance, perform CRUD operations, run aggregation pipelines, and manage indexes — all using Python syntax. It's the standard bridge between Python data/ML workflows and MongoDB storage.

10. ACID properties in the context of MongoDB transactions

- Atomicity — all operations in a transaction succeed or all are rolled back

   Consistency — data moves from one valid state to another

   Isolation — concurrent transactions don't see each other's intermediate state

   Durability — committed transactions persist even after a crash

11. Purpose of MongoDB's explain() function
- explain() is a query analysis tool. It shows how MongoDB executes a query — whether it used an index or did a full scan, how many documents were examined vs returned, and execution time. It's the primary tool for query optimization and diagnosing slow queries.

 Example = db.orders.find({ region: "West" }).explain("executionStats")

12. How MongoDB handles schema validation
- MongoDB supports schema validation using JSON Schema rules defined at the collection level with $jsonSchema. You can enforce field types, required fields, value ranges, and patterns. Validation mode can be set to strict (rejects all invalid docs) or moderate (only validates existing docs on update). This bridges the gap between full flexibility and structured data requirements.

13. Primary vs Secondary node in a replica set
 - The primary node receives all write operations and is the authoritative source of data. Secondary nodes continuously replicate the primary's oplog (operations log) to stay in sync. Secondaries can serve read operations (if configured) to reduce primary load. If the primary goes down, secondaries elect a new primary automatically.

15. Embedded documents and when to use them
- An embedded document is a document nested inside another document as a field value. For example, storing a patient's address as a sub-document inside the patient record rather than in a separate collection.
Use embedded documents when:

The nested data is always accessed together with the parent (e.g. order + order items)
The relationship is one-to-few (not one-to-millions)
You want to avoid JOINs for performance

16. What is the purpose of MongoDB’s $lookup stage in aggregation?

- performs a left outer join between two collections — similar to SQL's JOIN. It matches documents from a foreign collection based on a specified field and adds the matched documents as an array field.

18. What are the advantages of using MongoDB for horizontal scaling?
- MongoDB's built-in sharding splits data across multiple servers using a shard key. Benefits include: no single point of storage bottleneck, commodity hardware can be added as data grows, read/write load is distributed across shards, and it handles petabyte-scale data that no single machine could hold. This is far more cost-effective than vertical scaling (buying bigger and bigger machines).

19.How do MongoDB transactions differ from SQL transactions?
- transactions SQL databases were built with ACID transactions as a core feature from day one — they are cheap, fast, and work across any tables. In MongoDB, single-document operations are always atomic, but multi-document transactions (added in v4.0) carry more overhead because MongoDB was originally optimized for document-level atomicity. MongoDB transactions are best used sparingly, for cases that genuinely need multi-collection atomicity — not as a default pattern like in SQL.

22. How can you secure access to a MongoDB database?

- Enable authentication (disabled by default in local installs — must be turned on)

- Create users with the minimum required roles (principle of least privilege)

- Use TLS/SSL for all client-server connections

- Restrict network access via bindIp to only trusted IPs

- Enable auditing for compliance tracking

- Regularly rotate credentials and review role assignments

- On Atlas, use VPC peering and IP access lists

23.  What is MongoDB’s WiredTiger storage engine, and why is it important?

- WiredTiger has been MongoDB's default storage engine since v3.2. It provides document-level concurrency control (multiple clients can modify different documents simultaneously without blocking each other), compression (snappy/zlib/zstd — reduces disk usage by 60–80% vs MMAPv1), and support for encryption at rest. It replaced the older MMAPv1 engine and is the foundation for MongoDB's performance and scalability at scale.
